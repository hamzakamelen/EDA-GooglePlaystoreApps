#### **Observation of Size**
Three Kinds of Data in Size Column
1. Varies with device
2. Number + "M"
3. Number + " "


>- A/C to Observation There are three kinds of values in Size, Now we count these

```python
#  M values
df['Size'].loc[df['Size'].str.contains('M')].value_counts().sum()
# Output: 8829
```

```python
#  k values
df['Size'].loc[df['Size'].str.contains('k')].value_counts().sum()
# Output: 316
```

```python
#  Varies with device values
df['Size'].loc[df['Size'].str.contains('Varies with device')].value_counts().sum()
# Output: 1695
```

#### Check again, if it only contain three kinds of values?

```python
len(df) == 8829+1695+316
# Output True
```

Now we create two Columns called Size in Bytes and Size in MB's

```python
def convert_size(size):
    if isinstance(size,str):
        if 'k' in size:
            return float(size.replace('k','')) *1024
        elif 'M' in size:
            return float(size.replace('M',''))*(1024*1024)
        elif 'Varies with device' in size:
            return np.nan
    return size
```

```python
df['Size'] = df['Size'].apply(convert_size)
```

```python
df['Size']
```

Data Converted into Bytes

```python
df.rename(columns={'Size':'Size_Bytes'},inplace=True)
df.head()
```

```python
# Convert Size in MB and create a new column
df['Size_MB'] = df['Size_Bytes'].apply(lambda x: x/(1024*1024))
```

```python
df['Size_KB'] = df['Size_Bytes'].apply(lambda x: x/(1024))
```

## **Impute Rating Null Values**

```python
# Which Install Category Most Depend on Rating
df.groupby('Install_Category')['Rating'].describe()
```

### **Impute with respect to Install_Category**

```python
mean_rating = df.groupby('Install_Category')['Rating'].mean()
```

```python
def impute_rating(row):
    if pd.isna(row['Rating']):
        return mean_rating.get(row['Install_Category'], row['Rating'])
    return row['Rating']

# Apply the imputation function
df['Rating'] = df.apply(impute_rating, axis=1)
```

```python
df.isnull().sum()
```

#### **Impute Rating of No Installation Category with 0.0**

```python
df['Rating'].isnull().sum()
df['Rating']=df['Rating'].fillna(0.0)
```

```python
df['Rating'].isnull().sum()
```

```python
df.isnull().sum()
```

#### **Now Impute Size_Bytes Size_KB and Size_MB**

```python
df['Size_Bytes'].min()
df['Size_Bytes'] = df['Size_Bytes'].fillna(df['Size_Bytes'].min())
```

```python
df.isnull().sum()
```

```python
df['Size_MB'].min()
df['Size_MB'] = df['Size_MB'].fillna(df['Size_MB'].min())
```

```python
df['Size_KB'].min()
df['Size_KB'] = df['Size_KB'].fillna(df['Size_KB'].min())
```
